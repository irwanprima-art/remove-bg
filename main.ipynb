```python
import os
import torch
from PIL import Image
from tqdm import tqdm
from transparent_background import Remover
```

```python
# Verify GPU
print("CUDA available:", torch.cuda.is_available())
print("GPU:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "CPU")
```

```python
INPUT_DIR = "images"
OUTPUT_DIR = "remove-bg"

os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
# Initialize background remover (automatically uses GPU if available)
remover = Remover()
```

```python
# Process all images
image_files = sorted([
    f for f in os.listdir(INPUT_DIR)
    if f.lower().endswith((".jpg", ".jpeg", ".png"))
])

for idx, filename in enumerate(tqdm(image_files), start=1):
    input_path = os.path.join(INPUT_DIR, filename)
    output_path = os.path.join(OUTPUT_DIR, f"image_{idx}.png")
    
    img = Image.open(input_path).convert("RGB")
    result = remover.process(img, type='rgba')  # Returns RGBA with transparent background
    result.save(output_path)

print("✅ Semua gambar berhasil diproses!")
```
